```python
columns = data.columns
columns
```

### Define Y(dependent variable) and X(independent variable)

```python
y = data['class_label']
print(y.shape,'\n')
y
```

```python
x = data[['class_name', 'alcohol', 'malic_acid', 'ash',
       'alcalinity_of_ash', 'magnesium', 'total_phenols', 'flavanoids',
       'nonflavanoid_phenols', 'proanthocyanins', 'color_intensity', 'hue',
       'od280', 'proline']]
x.shape
```

#### or u can also use drop fun. to

```python
x = data.drop(['class_label', 'class_name'], axis = 1)
print(x.shape, '\n')
x
```

### Train nd Test Dataset

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size= 0.8, random_state= 777)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
x_test
```

### Train the Model using LinearRegression

```python
model= LinearRegression()
```

```python
model.fit(x_train, y_train)
```

### Predict the y_test value


```python
y_pred = model.predict(x_test)
print(y_pred.shape, '\n')
y_pred
```

### Model Evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, r2_score
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

### Visualization of Actual Vs Predicted Value

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel("Actual prices")
plt.ylabel("Predicted prices")
plt.title("Actual vs Predicted Prices")
plt.show()
```
